何を抽出するかだが、とりあえず、テキスト文字を抽出してみよう。
ファイル構造を見てみると、どうやらテキストの場合、最後に　\n　が付いている。これをパターンにしてみよう。

最初、上手く動かなかったので、色々調べてみた。特に、matchオブジェクトがよくわからなかったので苦労した。それでも、下記を読んで試したら、上手くいった。
https://docs.python.org/ja/3/library/re.html#match-objects


```python
import re

fp = open("about decorator.ipynb","r")

pattern = re.compile(r'.*\\n')
#pattern = re.compile(r'^(#+) (.+)$')

for line in fp:
        match = pattern.search(line)
        # (見出しレベル, 見出しのテキスト) の形で返す
        if match:
            print(match.group(0))
#            print(match.group(1))
#            print(match.group(2))
#            headings.append((len(match.group(1)), match.group(2)))
fp.close()
```
